```python
# Evaluation single sample halueval, triviaqa, mmlu
pipeline = data_generation.Pipeline()
halueval_data = loaders.load_halueval(sample_size=5000)
triviaqa_data = loaders.load_triviaqa(sample_size=5000)
mmlu_data = loaders.load_mmlu(sample_size=5000)
```
